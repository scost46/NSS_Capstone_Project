```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df1 = pd.read_csv('../data/USA_obesity_prevelance_since1975.csv')
```

```python
df1
```

```python
usa_row = df1.rename(columns={'Unnamed: 0': 'countries'})
```

```python
usa_row
```

```python
usa_over_time = usa_row[usa_row['countries'] == 'United States of America']
```

```python
usa_over_time
```

```python
def modify_column_names(column):
    if column.endswith('.1'):
        year = column.split('.')[0]
        return year + ' Male'
    elif column.endswith('.2'):
        year = column.split('.')[0]
        return year + ' Female'
    else:
        return column + ' Both Sexes'

# Modify the column names using the function
usa_over_time_clean = usa_over_time.rename(columns=modify_column_names)

# Print the filtered row with the modified column names
usa_over_time_clean
```

```python
columns_to_keep = [col for col in usa_over_time_clean.columns if 'Both Sexes' in col]
usa_over_time_clean = usa_over_time_clean[columns_to_keep]
```

```python
usa_over_time_clean
```

```python
usa_over_time_clean = usa_over_time_clean.rename(columns={'countries Both Sexes': 'USA'})
usa_over_time_clean
```

```python
usa_over_time_clean = usa_over_time_clean.rename(columns=lambda x: x.split(' ')[0])
```

```python
usa_over_time_clean
```

```python
columns_to_clean = usa_over_time_clean.columns

# Remove the values within brackets in the specified columns
usa_over_time_clean[columns_to_clean] = usa_over_time_clean[columns_to_clean].apply(lambda x: x.str.replace(r'\[.*?\]', '', regex=True))
```

```python
usa_over_time_clean
```

```python
sorted_years_df = usa_over_time_clean.iloc[:, np.argsort(usa_over_time_clean.columns)]
sorted_years_df
```

```python
sorted_years_df = sorted_years_df.drop('USA', axis=1)
sorted_years_df
```

```python
sorted_years_df = sorted_years_df.rename(index={189: 'USA'})
sorted_years_df
```
