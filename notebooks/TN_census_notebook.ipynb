```python
# Get the column names
column_names = TN_demo.columns.tolist()

# a list to store the columns to be dropped
columns_to_drop = []

# Iterate over the column names
for i, column in enumerate(column_names):
    # Check if the column is unnamed and followed by an even number
    if column.startswith("Unnamed:") and (i + 2) % 2 == 0:
        columns_to_drop.append(column)

# Drop the specified columns
TN_demo = TN_demo.drop(columns_to_drop, axis=1)
```

```python
TN_demo
```

```python
# Store the label column
label_column = TN_demo.columns[0]

# Iterate over the column names starting from the second column
for i in range(1, len(TN_demo.columns)):
    column = TN_demo.columns[i]
    # Check if the column is unnamed and the previous column is not unnamed
    if column.startswith("Unnamed:") and not TN_demo.columns[i-1].startswith("Unnamed:"):
        # Get the county name from the previous column
        county_name = TN_demo.columns[i-1]
        # Replace the unnamed column name with the county name
        TN_demo = TN_demo.rename(columns={column: county_name})

# Drop the unnamed columns except for the first column
TN_demo = TN_demo.drop(TN_demo.columns[TN_demo.columns.str.startswith("Unnamed:")], axis=1)
```

```python
TN_demo
```

```python
# Drop rows with "Male" and "Female" values in the label column
TN_demo = TN_demo[~TN_demo['Label'].isin(['Male', 'Female'])]
TN_demo
```

```python
# Define the rows to be dropped
rows_to_drop = [1,34, 42, 43, 44, 45, 55, 56, 57, 58, 65, 73, 91]

# Drop the specified rows
TN_demo = TN_demo.drop(rows_to_drop)

# Reset the index
TN_demo = TN_demo.reset_index(drop=True)
TN_demo
```

```python

# Get the current column titles
current_columns = TN_demo.columns.tolist()

# Get the values from the first row
first_row_values = TN_demo.iloc[0].tolist()

# Combine the column titles and first row values
new_columns = [f"{title} {value}" for title, value in zip(current_columns, first_row_values)]

# Assign the new column titles to the dataframe
TN_demo.columns = new_columns

# Remove the first row since it is now incorporated in the column titles
TN_demo = TN_demo.iloc[1:]

# Reset the index
TN_demo.reset_index(drop=True, inplace=True)
```

```python
# Create a new DataFrame with the new structure
organized_df = pd.DataFrame(columns=['Label', 'County', 'Value'])

# Iterate over the rows of the original DataFrame
for index, row in new_income_df.iterrows():
    label = row['Label']
    
    # Iterate over the columns (counties) of the original DataFrame
    for column in new_income_df.columns[1:]:
        county = column.split(',')[0].strip()
        value = row[column]
        
        # Append a new row to the organized DataFrame
        organized_df = organized_df.append({'Label': label, 'County': county, 'Value': value}, ignore_index=True)
```

```python
organized_df
```

```python

# Remove commas and percentage symbols from the "Value" column
organized_df['Value'] = organized_df['Value'].str.replace(',', '').str.replace('%', '')

# Convert the "Value" column to numeric
organized_df['Value'] = pd.to_numeric(organized_df['Value'])

# Verify the data types
print(organized_df.dtypes)
```
