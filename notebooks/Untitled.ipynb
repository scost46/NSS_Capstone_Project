```python
import pandas as pd
```

```python
locations = pd.read_csv('../data/top_food_chains_in_tennessee_usa.csv')
```

```python
columns_to_drop = ['Phone', 'Open Hours', 'Updated Date', 'Country']
locations_cleaned = locations.drop(columns=columns_to_drop)
locations_cleaned
```

```python
counties_to_keep = [
    "Anderson", "Blount", "Bradley", "Davidson", "Greene", "Hamilton",
    "Knox", "Madison", "Maury", "Montgomery", "Putnam", "Robertson",
    "Rutherford", "Sevier", "Shelby", "Sullivan", "Sumner", "Washington",
    "Williamson", "Wilson"
]

# Filtering the DataFrame based on the specified counties
census_locations = locations_cleaned[locations_cleaned['County'].isin(counties_to_keep)].copy()

# Resetting the index of the new DataFrame
census_locations.reset_index(drop=True, inplace=True)
```
